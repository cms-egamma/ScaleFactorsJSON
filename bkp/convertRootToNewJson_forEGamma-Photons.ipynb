```python
corr = Correction.parse_obj(
    {
        "version": 2,
        "name": "UL-Photon-ID-SF",
        "description": "Photon ID Scale Factors for Ultra Legacy",
        "inputs": [
            {"name": "year","type": "string", "description": "year/scenario: example 2016preVFP, 2017 etc"},
            {"name": "ValType","type": "string", "description": "sf or syst (Currently syst is total syst) "},
            {"name": "WorkingPoint","type": "string", "description": "Working Point of choice : Loose, Medium etc."},
            {"type": "real", "name": "eta", "description": "supercluster eta"},
            {"name": "pt", "type": "real", "description": "photon pT"},
        ],
        "output": {"name": "weight", "type": "real", "description": "value of scale factor or systematic uncertainty"},
        "data": schema.Category.parse_obj({
            "nodetype": "category",
            "input": "year",
            "content": [
                schema.CategoryItem.parse_obj({"key":"2017",
                                               "value": SFyearwise(files=files2017,names=namesSFs)}),
                schema.CategoryItem.parse_obj({"key":"2018",
                                               "value": SFyearwise(files=files2018,names=namesSFs)}),
                schema.CategoryItem.parse_obj({"key":"2016preVFP",
                                               "value": SFyearwise(files=files2016preVFP,names=namesSFs)}),
                schema.CategoryItem.parse_obj({"key":"2016postVFP",
                                               "value": SFyearwise(files=files2016postVFP,names=namesSFs)}),]
        })
    })
        
corrs.append(corr)

corr = Correction.parse_obj(
    {
        "version": 2,
        "name": "UL-Photon-PixVeto-SF",
        "description": "Photon Pixel Veto Scale Factors for Ultra Legacy",
        "inputs": [
            {"name": "year","type": "string", "description": "year/scenario: example 2016preVFP, 2017 etc"},
            {"name": "ValType","type": "string", "description": "sf or syst (Currently syst is total syst) "},
            {"name": "WorkingPoint","type": "string", "description": "Working Point of choice : Loose, Medium etc."},
            {"type": "string", "name": "HasPixBin", "description": "Possible choices: ['EBInc','EBHighR9','EBLowR9','EEInc','EEHighR9','EELowR9']"},
        ],
        "output": {"name": "weight", "type": "real", "description": "value of scale factor or systematic uncertainty"},
        "data": schema.Category.parse_obj({
            "nodetype": "category",
            "input": "year",
            "content": [
                schema.CategoryItem.parse_obj({"key":"2017",
                                               "value": HasPixSFyearwise(files=HasPixfiles2017,names=HasPixnames)}),
                schema.CategoryItem.parse_obj({"key":"2018",
                                               "value": HasPixSFyearwise(files=HasPixfiles2018,names=HasPixnames)}),
                schema.CategoryItem.parse_obj({"key":"2016preVFP",
                                               "value": HasPixSFyearwise(files=HasPixfiles2016preVFP,names=HasPixnames)}),
                schema.CategoryItem.parse_obj({"key":"2016postVFP",
                                               "value": HasPixSFyearwise(files=HasPixfiles2016postVFP,names=HasPixnames)}),]
        })
    })

corrs.append(corr)

corr = Correction.parse_obj(
    {
        "version": 2,
        "name": "UL-Photon-CSEV-SF",
        "description": "Photon CSEV Scale Factors for Ultra Legacy",
        "inputs": [
            {"name": "year","type": "string", "description": "year/scenario: example 2016preVFP, 2017 etc"},
            {"name": "ValType","type": "string", "description": "sf or syst (Currently syst is total syst) "},
            {"name": "WorkingPoint","type": "string", "description": "Working Point of choice : Loose, Medium etc."},
            {"type": "string", "name": "CSEVBin", "description": "Possible choices: ['EBInc','EBHighR9','EBLowR9','EEInc','EEHighR9','EELowR9']"},
        ],
        "output": {"name": "weight", "type": "real", "description": "value of scale factor or systematic uncertainty"},
        "data": schema.Category.parse_obj({
            "nodetype": "category",
            "input": "year",
            "content": [
                schema.CategoryItem.parse_obj({"key":"2017",
                                               "value": CSEVSFyearwise(files=CSEVfiles2017,names=CSEVnames)}),
                schema.CategoryItem.parse_obj({"key":"2018",
                                               "value": CSEVSFyearwise(files=CSEVfiles2018,names=CSEVnames)}),
                schema.CategoryItem.parse_obj({"key":"2016preVFP",
                                               "value": CSEVSFyearwise(files=CSEVfiles2016preVFP,names=CSEVnames)}),
                schema.CategoryItem.parse_obj({"key":"2016postVFP",
                                               "value": CSEVSFyearwise(files=CSEVfiles2016postVFP,names=CSEVnames)}),]
        })
    })

corrs.append(corr)


#Save JSON
from correctionlib.schemav2 import CorrectionSet
import gzip

cset = CorrectionSet(schema_version=2, corrections=corrs)

with open(nameJSON, "w") as fout:
    fout.write(cset.json(exclude_unset=True, indent=4))
```

```python
#Evaluator Example

from correctionlib import _core
evaluator = _core.CorrectionSet.from_string(cset.json(exclude_unset=True, indent=4))
```
